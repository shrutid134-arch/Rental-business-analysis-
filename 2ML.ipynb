```python
# pip install if you need them (uncomment to run)
# !pip install sqlalchemy psycopg2-binary pandas scikit-learn matplotlib joblib

import pandas as pd
import numpy as np
from sqlalchemy import create_engine, text
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import classification_report, roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt
import joblib   # to save models

# Display settings for notebooks
pd.set_option('display.max_columns', 200)
```

```python
# Example: create a customer-level features table from payments and rentals.
# Adjust table names if your schema differs (you mentioned adv_top_customers, customer_segmentation etc).
query = """
WITH payments_agg AS (
    SELECT
      customer_id,
      COUNT(*) AS n_payments,
      SUM(amount) AS total_spend,
      AVG(amount) AS avg_payment,
      MAX(payment_date) AS last_payment_date,
      MIN(payment_date) AS first_payment_date
    FROM payment
    GROUP BY customer_id
),
rentals_agg AS (
    SELECT
      customer_id,
      COUNT(*) AS n_rentals,
      MAX(rental_date) AS last_rental_date,
      MIN(rental_date) AS first_rental_date
    FROM rental
    GROUP BY customer_id
)
SELECT
  c.customer_id,
  c.store_id,
  COALESCE(p.total_spend, 0) AS total_spend,
  COALESCE(p.n_payments, 0) AS n_payments,
  COALESCE(p.avg_payment, 0) AS avg_payment,
  COALESCE(r.n_rentals, 0) AS n_rentals,
  GREATEST(
    COALESCE(EXTRACT(EPOCH FROM (now() - COALESCE(p.last_payment_date, r.last_rental_date))) / 86400, 9999),
    9999
  ) AS recency_days,
  COALESCE(p.last_payment_date, r.last_rental_date) AS last_activity
FROM customer c
LEFT JOIN payments_agg p ON c.customer_id = p.customer_id
LEFT JOIN rentals_agg r ON c.customer_id = r.customer_id
;
"""

df_customers = pd.read_sql(query, engine)
# quick clean
df_customers['recency_days'] = df_customers['recency_days'].fillna(9999)
df_customers.head()
```

```python
# 3.1 Prepare RFM features
rfm = df_customers[['customer_id', 'total_spend', 'n_payments', 'n_rentals', 'recency_days']].copy()
rfm_features = rfm[['total_spend', 'n_payments', 'n_rentals', 'recency_days']].fillna(0)

scaler = StandardScaler()
X_rfm = scaler.fit_transform(rfm_features)

# 3.2 Find best k using silhouette (try k=2..6)
sil_scores = {}
for k in range(2, 7):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_rfm)
    sil = silhouette_score(X_rfm, labels)
    sil_scores[k] = sil
    print(f"k={k}, silhouette={sil:.4f}")

# Choose k with highest silhouette
best_k = max(sil_scores, key=sil_scores.get)
print("Best k by silhouette:", best_k)

# 3.3 Fit final KMeans
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
rfm['cluster'] = kmeans.fit_predict(X_rfm)

# 3.4 Inspect cluster centers (in original scale)
centers = scaler.inverse_transform(kmeans.cluster_centers_)
centers_df = pd.DataFrame(centers, columns=rfm_features.columns)
centers_df['cluster'] = centers_df.index
centers_df = centers_df[['cluster', 'total_spend', 'n_payments', 'n_rentals', 'recency_days']]
print("\nCluster centers (original scale):")
display(centers_df)

# 3.5 Attach cluster to the original df
df_customers = df_customers.merge(rfm[['customer_id','cluster']], on='customer_id', how='left')

# 3.6 Simple cluster counts
print(df_customers['cluster'].value_counts().sort_index())
```

```python
# Summarize clusters for quick talking points
cluster_summary = df_customers.groupby('cluster').agg(
    customers = ('customer_id', 'count'),
    avg_spend = ('total_spend', 'mean'),
    avg_payments = ('n_payments', 'mean'),
    avg_rentals = ('n_rentals', 'mean'),
    avg_recency = ('recency_days', 'mean')
).reset_index().sort_values('cluster')

cluster_summary
```

```python
# Label: churn = 1 if last_activity older than 90 days
THRESHOLD_DAYS = 90
df_customers['last_activity'] = pd.to_datetime(df_customers['last_activity'])
df_customers['days_since_last'] = (pd.Timestamp.now(tz=df_customers['last_activity'].dt.tz) - df_customers['last_activity']).dt.days.fillna(9999)
df_customers['churn'] = (df_customers['days_since_last'] > THRESHOLD_DAYS).astype(int)

df_customers[['customer_id','last_activity','days_since_last','churn']].head()
```

```python
# Use simple, interpretable features
features = df_customers[[
    'total_spend', 'n_payments', 'n_rentals', 'avg_payment', 'recency_days', 'store_id'
]].copy()

# handle missing
features = features.fillna(0)

# Optionally one-hot encode store_id if few stores
features = pd.get_dummies(features, columns=['store_id'], prefix='store', drop_first=True)

# Target
y = df_customers['churn']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.25, random_state=42, stratify=y)

X_train.shape, X_test.shape, y_train.mean(), y_test.mean()
```

```python
# Train
dt = DecisionTreeClassifier(max_depth=5, min_samples_leaf=20, random_state=42)  # shallow & interpretable
dt.fit(X_train, y_train)

# Text rule export (very interview-friendly)
tree_rules = export_text(dt, feature_names=list(X_train.columns))
print("Decision tree rules (text):\n")
print(tree_rules)

# Predictions and metrics
y_pred = dt.predict(X_test)
y_proba = dt.predict_proba(X_test)[:,1]

print("\nClassification report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)
```

```python
# --- Balanced churn definition using inactivity percentile ---

# Calculate days since last activity
df_customers['last_activity'] = pd.to_datetime(df_customers['last_activity'])
df_customers['days_since_last'] = (
    pd.Timestamp.now(tz=df_customers['last_activity'].dt.tz) 
    - df_customers['last_activity']
).dt.days.fillna(9999)

# Define churn threshold as 70th percentile (top 30% most inactive customers)
churn_threshold = df_customers['days_since_last'].quantile(0.70)

# Label churn = 1 if customer is more inactive than threshold
df_customers['churn'] = (df_customers['days_since_last'] > churn_threshold).astype(int)

# Check distribution
print("Churn distribution:\n", df_customers['churn'].value_counts(normalize=True).round(2))
```

```python
# ============================
# Customer Churn Prediction (Decision Tree)
# ============================

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Balanced churn label (top 30% most inactive customers)
df_customers['last_activity'] = pd.to_datetime(df_customers['last_activity'])
df_customers['days_since_last'] = (
    pd.Timestamp.now(tz=df_customers['last_activity'].dt.tz) 
    - df_customers['last_activity']
).dt.days.fillna(9999)

churn_threshold = df_customers['days_since_last'].quantile(0.70)
df_customers['churn'] = (df_customers['days_since_last'] > churn_threshold).astype(int)

print("Churn distribution:\n", df_customers['churn'].value_counts(normalize=True).round(2))

# 2. Features + Target
features = df_customers[[
    'total_spend', 'n_payments', 'n_rentals', 'avg_payment', 'recency_days', 'store_id'
]].fillna(0)

# One-hot encode store_id
features = pd.get_dummies(features, columns=['store_id'], prefix='store', drop_first=True)

y = df_customers['churn']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    features, y, test_size=0.25, random_state=42, stratify=y
)

print("\nTrain distribution:\n", y_train.value_counts(normalize=True).round(2))
print("Test distribution:\n", y_test.value_counts(normalize=True).round(2))

# 3. Train Decision Tree
dt = DecisionTreeClassifier(max_depth=5, min_samples_leaf=20, 
                            class_weight="balanced", random_state=42)
dt.fit(X_train, y_train)

# Show tree rules
tree_rules = export_text(dt, feature_names=list(X_train.columns))
print("\nDecision tree rules:\n")
print(tree_rules)

# 4. Predictions + Metrics
y_pred = dt.predict(X_test)
y_proba = dt.predict_proba(X_test)[:,1]

print("\nClassification report:")
print(classification_report(y_test, y_pred))

print("\nConfusion matrix:")
print(confusion_matrix(y_test, y_pred))

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {roc_auc:.3f})')
plt.plot([0,1],[0,1],'--', linewidth=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()

# 5. Feature Importance
importances = pd.Series(dt.feature_importances_, index=X_train.columns).sort_values(ascending=False)
print("\nTop features:\n", importances.head(10))

importances.head(10).plot(kind='bar', figsize=(8,4))
plt.title("Feature Importances (Decision Tree)")
plt.tight_layout()
plt.show()

# 6. Top 100 At-Risk Customers
df_scores = X_test.copy()
df_scores['customer_id'] = df_customers.loc[X_test.index, 'customer_id'].values
df_scores['churn_proba'] = y_proba
top_100_at_risk = df_scores.sort_values('churn_proba', ascending=False).head(100)

print("\nTop 10 at-risk customers:")
print(top_100_at_risk[['customer_id','churn_proba']].head(10))
```
